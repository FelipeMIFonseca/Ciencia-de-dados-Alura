# laços e loops

```python
#for fora da função

def verificar_idade(idade):
  if idade >= 18:
    print(f'{idade} anos de idade, TEM permissão para dirigir')
  else:
    print(f'{idade} anos de idade, NÃO TEM permissão para dirigir')
for idade in idades:
  verificar_idade(idade)
```

```python
#for dentro da função

def verificar_com_for_dentro(idades):
  for idade in idades:
    if idade >= 18:
     print(f'{idade} anos de idade, TEM permissão para dirigir')
    else:
     print(f'{idade} anos de idade, NÃO TEM permissão para dirigir')
verificar_com_for_dentro(idades)
```

```python
permissoes = []
idades = [21, 41, 51, 61, 13, 14, 15, 16, 17]

def verificar_as_idades(idades, permissoes):
  for idade in idades:
    if idade >= 18:
      permissoes.append(True)
    else:
      permissoes.append(False)

verificar_as_idades(idades, permissoes)

permissoes
```

```python
for permissao in permissoes:
  if permissao == True:
    print('permissão para dirigir')
  else:
    print('não tem permissão para dirigir')
```
